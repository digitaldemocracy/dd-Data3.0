```python
def check_overlaps(row, cpy_leg_lop_df):
    out = False
    matched_rows = cpy_leg_lop_df[(cpy_leg_lop_df.pid == row['pid']) &
                                  (row.name != cpy_leg_lop_df.index) &
                                  (row['hire_date'] <= cpy_leg_lop_df.hire_date) &
                                  (row['end_date'] >= cpy_leg_lop_df.hire_date)]

    if len(matched_rows.index):
        out = True
    matched_rows = cpy_leg_lop_df[(cpy_leg_lop_df.pid == row['pid']) &
                                  (row.name != cpy_leg_lop_df.index) &
                                  (row['hire_date'] <= cpy_leg_lop_df.end_date) &
                                  (row['end_date'] >= cpy_leg_lop_df.end_date)]

    if len(matched_rows.index):
        out = True
    matched_rows = cpy_leg_lop_df[(cpy_leg_lop_df.pid == row['pid']) &
                                  (row.name != cpy_leg_lop_df.index) &
                                  (row['hire_date'] >= cpy_leg_lop_df.hire_date) &
                                  (row['end_date'] <= cpy_leg_lop_df.end_date)]

    if len(matched_rows.index):
        out = True
    return out
    
```
